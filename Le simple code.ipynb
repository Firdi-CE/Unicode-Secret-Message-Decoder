```python
import requests
from bs4 import BeautifulSoup

def print_google_doc_grid(url: str):

    # Step 1: Fetch the HTML content from the URL
    try:
        response = requests.get(url)
        # Raise an error for bad responses (e.g., 404, 500)
        response.raise_for_status()
        html_content = response.text
    except requests.exceptions.RequestException as e:
        print(f"Error: Could not fetch URL. {e}")
        return

    # Step 2: Parse the HTML to find the data
    soup = BeautifulSoup(html_content, 'html.parser')

    data_points = []
    max_x = 0
    max_y = 0

    # Find all table rows
    rows = soup.find_all('tr')

    for row in rows:
        cells = row.find_all('td')

        # checks if it's a data row with exactly 3 columns
        if len(cells) == 3:
            try:
                # Extract the text from each cell
                x_str = cells[0].get_text(strip=True)
                char = cells[1].get_text(strip=True)
                y_str = cells[2].get_text(strip=True)

                # Convert coordinates to integers
                x = int(x_str)
                y = int(y_str)

                # If character cell is empty, use a space
                if not char:
                    char = ' '

                data_points.append((x, y, char))

                # checks the maximum coordinates to define the grid size
                max_x = max(max_x, x)
                max_y = max(max_y, y)

            except ValueError:
                # skips rows that doesn't have integers

                pass
            except Exception as e:
                print(f"Warning: Skipping row due to parse error: {e}")

    if not data_points:
        print("No valid data points were found in the document.")
        return

    # Step 3: Create the 2D grid
    num_cols = max_x + 1
    num_rows = max_y + 1

    # Initialize the grid with spaces
    grid = [[' ' for _ in range(num_cols)] for _ in range(num_rows)]

    # Step 4: Populate the grid
    for x, y, char in data_points:
        grid_row = max_y - y  # Invert the y-axis
        grid_col = x

        grid[grid_row][grid_col] = char

    # Step 5: Print the grid
    for row in grid:
        print("".join(row))
```

```python
# The URL from the prompt
doc_url = "https://docs.google.com/document/d/e/2PACX-1vSZ9d7OCd4QMsjJi2VFQmPYLebG2sGqI879_bSPugwOo_fgRcZLAFyfajPWU91UDiLg-RxRD41lVYRA/pub"

# Call the function
print_google_doc_grid(doc_url)
```
